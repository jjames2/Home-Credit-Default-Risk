```python
import pandas as pd
import numpy as np

import datetime
```

```python
application_train = pd.read_csv('./DATA/application_train.csv')
application_train.head()
```

```python
application_train['DEF_30_RATIO_SOCIAL_CIRCLE'] = application_train['DEF_30_CNT_SOCIAL_CIRCLE'] / application_train['OBS_30_CNT_SOCIAL_CIRCLE']
application_train['DEF_30_RATIO_SOCIAL_CIRCLE'].fillna(0, inplace=True)
application_train['DEF_60_RATIO_SOCIAL_CIRCLE'] = application_train['DEF_60_CNT_SOCIAL_CIRCLE'] / application_train['OBS_60_CNT_SOCIAL_CIRCLE']
application_train['DEF_60_RATIO_SOCIAL_CIRCLE'].fillna(0, inplace=True)
```

```python
application_train.head()
```

```python
application_train['AMT_REQ_CREDIT_BUREAU'] = application_train['AMT_REQ_CREDIT_BUREAU_HOUR'] + application_train['AMT_REQ_CREDIT_BUREAU_DAY'] + application_train['AMT_REQ_CREDIT_BUREAU_WEEK'] + application_train['AMT_REQ_CREDIT_BUREAU_MON'] + application_train['AMT_REQ_CREDIT_BUREAU_QRT'] + application_train['AMT_REQ_CREDIT_BUREAU_YEAR']
```

```python
application_train.head()
```

```python
credit_reqs_cols = [col for col in application_train.columns if 'AMT_REQ_CREDIT_BUREAU' in col]
for col in credit_reqs_cols:
    application_train[col].fillna(0, inplace=True)
```

```python
application_train.head()
```

### Fill nulls

```python
application_train['NAME_TYPE_SUITE'].fillna('Unaccompanied', inplace=True)
```

```python
application_train['CNT_FAM_MEMBERS'].fillna(application_train['CNT_FAM_MEMBERS'].mean(), inplace=True)
```

```python
application_train['OCCUPATION_TYPE'].fillna('Unknown', inplace=True)
```

```python
social_circle_cols = [col for col in application_train.columns if 'SOCIAL_CIRCLE' in col]
for col in social_circle_cols:
    application_train[col].fillna(0, inplace=True)
```

```python
application_train['DAYS_LAST_PHONE_CHANGE'].fillna(0, inplace=True)
```

```python
rem_cols = [col for col in application_train.columns if 'AVG' in str(col) or 'MEDI' in str(col) or 'MODE' in str(col)]
add_rem_cols = ['OWN_CAR_AGE']
rem_cols.extend(add_rem_cols)
len(rem_cols)
```

### Fill category values representing null with actual null value

```python
tf = application_train['CODE_GENDER'] == 'XNA'
application_train.loc[tf, 'CODE_GENDER'] = 'M'
```

```python
tf = application_train['NAME_FAMILY_STATUS'] == 'Unknown'
application_train.loc[tf, 'NAME_FAMILY_STATUS'] = 'Married'
```

### Drop bad columns

```python
keep_cols = [col for col in application_train.columns if str(col) not in rem_cols]
final_dataset = application_train[keep_cols].copy()
```

```python
final_dataset.head()
```

### Add default count from bureau balances

```python
applicant_default_count = pd.read_csv('./DATA/applicant_default_count.csv')
applicant_default_count.head()
```

```python
rcols = ['SK_ID_CURR','DEFAULT_COUNT']
final_dataset2 = final_dataset.merge(applicant_default_count[rcols], on='SK_ID_CURR', how='left')
final_dataset2['DEFAULT_COUNT'].fillna(0, inplace=True)
final_dataset2.head()
```

```python
miss_cols = final_dataset2.columns[final_dataset2.isna().any()].tolist()
miss_cols
```

### KNN Imputation

```python
# interpolate add: DAYS_BIRTH, DEFAULT_COUNT
# academic degree should be higher edu
```

```python
imp_cols = ['SK_ID_CURR'] + ['AMT_ANNUITY','AMT_GOODS_PRICE','EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3'] + ['AMT_INCOME_TOTAL','AMT_CREDIT','DAYS_BIRTH','DEFAULT_COUNT']
```

```python
impute_df = final_dataset2[imp_cols].copy()
impute_df.index = final_dataset2['SK_ID_CURR']
del impute_df['SK_ID_CURR']
impute_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler = MinMaxScaler()
impute_df_scaled = pd.DataFrame(scaler.fit_transform(impute_df),columns=impute_df.columns)
impute_df_scaled.head()
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
```

```python
len(impute_df_scaled)
```

```python
start = datetime.datetime.now()
imputer.fit(impute_df_scaled.sample(n=100000))
final_impute_df_scaled = pd.DataFrame(imputer.transform(impute_df_scaled),columns = impute_df_scaled.columns)
end = datetime.datetime.now()
tt = (end-start).seconds
print("Time taken in minutes: " + str(tt/60))
```

```python
# final_impute_df_scaled = final_impute_df.copy()
final_impute_df_scaled.head()
```

```python
final_impute_df = pd.DataFrame(scaler.inverse_transform(final_impute_df_scaled),columns=impute_df.columns)
final_impute_df.index = impute_df.index
final_impute_df.head()
```

```python
final_impute_df.reset_index(inplace=True)
```

```python
final_dataset2.to_csv('./DATA/imputed_data.csv')
```

```python
(final_dataset2['SK_ID_CURR']==final_impute_df['SK_ID_CURR']).sum()
```

```python
for col in miss_cols:
    final_dataset2[col] = final_impute_df[col]
```
